```python
import matplotlib.pyplot as plt
import numpy as np
```

# **MatPlotLib Basics**

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='sin(x)')
plt.title("Sine Wave")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Different Kinds of plots
fig, axes = plt.subplots(2, 2)
fig.set_size_inches(10, 10)

x = np.linspace(0, 10, 100)
y = np.sin(x)

y = np.sin(x) + np.random.randn(x.size)*0.2
axes[0, 0].scatter(x, y)
axes[0, 0].set_title("scatter")

axes[0, 1].hist(y, bins=10)
axes[0, 1].set_title("histogram")

axes[1, 0].semilogx(x, y)
axes[1, 0].set_title("log (along x)")

x = ['A', 'B', 'C']
y = [10, 20, 30]
axes[1, 1].bar(x, y)
axes[1, 1].set_title("bar")

plt.show()
```
